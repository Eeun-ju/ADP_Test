```python
plt.bar(data.grade.value_counts().index, data.grade.value_counts().values)
for i in range(len(data.grade.value_counts().index)):
  height = data.grade.value_counts().values[i]
  plt.text(data.grade.value_counts().index[i], height + 0.25, '%.1f'%height, ha = 'center', va = 'bottom', size = 8)
plt.title('Set Title ')
plt.xlabel('Grade')
plt.ylabel('Num of Studets')
```
