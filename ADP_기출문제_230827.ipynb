```python
import scipy.stats as stats
stats.shapiro(data.grade) # 문제에서 종속변수인 성적 - 성적이 어떤 분포인지 그래프로 확인할 수 있었음
# shapiro 함수는 데이터가 정규분포를 따르는지 검정하는 방법으로 귀무가설 : 정규분포 vs 대립가설 : 정규분포x
# 함수 실행 결과 p-value 0.0014로 유의수준 5%하, 귀무가설을 기각하며 (0.0014 < 0.05 이므로) 성적은 정규분포를 따르지 않는다고 할 수 있다.
```
